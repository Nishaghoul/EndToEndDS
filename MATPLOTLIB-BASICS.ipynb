```python
#CLASS TASK-VI
import pandas as pd
import numpy as np
df_iris=pd.read_csv(r'C:\Users\Lenovo\Anaconda3\pkgs\scikit-learn-0.21.3-py37h6288b17_0\Lib\site-packages\sklearn\datasets\data\\iris.csv')
x=df_iris['setosa']
y=df_iris['versicolor']
z=df_iris['virginica']
d=np.arange(150)
plt.plot(d,x)
plt.plot(d,y)
plt.plot(d,z)
plt.xlabel('Features')
plt.ylabel('Species')
plt.title('Iris Species')
plt.legend(('SETOSA','VERSICOLOR','VIRGINICA'))
plt.show()
```
